Si queremos hacer el análisis separando Hombre y Mujeres, primeros tenemos que filtrar. Una manera es usar `.loc`.
También se podria usar `.where()` y luego quitar 

La explicación para `.loc` es la siguiente: 

- Cuando usamos una columna en una condición, el resultado es una columna de valores booleanos (verdadero o falso, o True o False). Por ejemplo `plaza['Sexo'] == 'Mujer'` devolverá True para aquellas filas donde la columna `Sexo` tiene el valor `'Mujer'` y `False` si es Hombre. 
- `.loc` permite seleccionar sólo las posiciones que son `True`. De esa manera filtramos "sólo mujeres". 


Recuerden que pueden ver la ayuda con `shift + tab` (varias veces para ampliar)


```python
mujeres = plaza.loc[plaza['Sexo'] == 'Mujer']

# luego el grafico es igual, 
mujeres['juego'].value_counts().plot.pie(autopct='%1.0f%%')
```

```python
hombres = plaza.loc[plaza['Sexo'] == 'Hombre'] 
hombres['juego'].value_counts().plot.pie(autopct='%1.0f%%')
```
