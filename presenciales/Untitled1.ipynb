```python
import pandas as pd
```

```python
plaza = pd.read_csv('/home/tin/Descargas/plaza.csv')
```

```python
dataset = plaza[(plaza['Fecha de nacimiento'].str.contains('-') | 
                 plaza['Fecha de nacimiento'].str.contains('/')) & ~
                 plaza['Fecha de nacimiento'].str.contains('o')]
```

```python
dataset
```

```python
dataset = dataset.drop(dataset.columns[0], axis=1)
```

```python
select = dataset.Sexo[dataset.Sexo == 'Hombre'].sample(5)
dataset.Sexo[select.index] = 'masculino'
```

```python
select = dataset.Sexo[dataset.Sexo == 'Mujer'].sample(5)
dataset.Sexo[select.index] = 'femenino'
```

```python
dataset['horas de plaza/año'] = pd.Series(np.random.randint(48, size=dataset.size), dtype=np.float64).astype(str) 
dataset.update(dataset.sample(4)['horas de plaza/año'].apply(lambda x: x.replace('.', ',')))
dataset
```

```python
dataset.to_csv('dataset_sucio.csv', index=False)
```
