# Y ahora analicemos la inflación de mayo

```python
import pandas as pd
```

Vamos abriendo todos los datasets disponibles

```python
precios_abril = pd.read_csv('../proyecto_integrador/precios_abril20.csv')
precios_abril.head()
```

```python
precios_mayo = pd.read_csv('../proyecto_integrador/precios_mayo22.csv')
precios_mayo.head()
```

```python
sucursales = pd.read_csv('../proyecto_integrador/sucursales.csv')
sucursales.head()
```

```python
cadenas = pd.read_csv('../proyecto_integrador/cadenas.csv')
cadenas.head()
```

```python
productos = pd.read_csv('../proyecto_integrador/productos.csv')
productos.head()
```

Haremos el primer cruce. Queremos los precios de abril y de mayo para cada producto-sucursal (es decir, las filas donde coincidan **ambos** valores)

```python
# merge recibe dos dataset, que se definen como `left` (izquierda) y `right` (derecha). en este caso 
# por defecto estamos haciendo un cruce tipo "inner" que es el que sólo deja las filas cuando 
# todas las columnas del criterio de cruce (parámetro `on=`) coinciden

precios = pd.merge(precios_abril, precios_mayo, on=['sucursal_id', 'producto_id'])
precios.head()
```

Por defecto, el merge agrega `_x` e `_y` como sufijos a las columnas respectivas del dataset de la izquierda y la derecha que no forman parte de las clave. Para ser más descriptivos, podemos definir otros sufijos

```python
precios = pd.merge(precios_abril, precios_mayo, on=['sucursal_id', 'producto_id'], suffixes=('_abril', '_mayo'))
precios.head()
```

Sólo a fines de "elegancia", vamos a reordenar las columnas del dataset. 
Para eso simplemente hacemos un indizado de una lista de columnas, en el orden en que las queremos
y lo asignamos a una nueva variable (que puede ser la misma que la original para evitar duplicar informacion y ahorrar memoria)

```python
columnas = ['sucursal_id', 'producto_id', 'precio_abril', 'precio_mayo']
precios = precios[columnas]
precios.head()
```

```python
precios['precio_mayo']
```

Ahora necesitamos calcular el aumento entre el precio de abril y el de mayo. 
Es decir, crearemos una nueva columna que haga el cálculo en terminos porcentuales

$$
\%\ aumento = \frac{nuevo - original}{original} \cdot 100
$$


Si las dimensiones de los operandos son compatibles (por ejemplo, las columnas tienen la misma cantidad de filas), esta formula se puede aplicar directamente utilizando las columnas, e internamente cada nuevo valor se calculará a partir de los valores de la fila correspondiente. 


```python
precios['aumento'] = (precios['precio_mayo'] - precios['precio_abril']) * 100 / precios['precio_abril']
precios.head()
```

Si ordenamos en orde descendente, podemos ver cuales fueron los mayores aumentos en Córdoba

```python
precios.sort_values('aumento', ascending=False)[:50]
```

Cómo vemos, hay aumentos de hasta 260%! 
De cualquier no son representativos ya que están fuera del grueso de los aumentos más tipicos. 
Para visualizar esto, lo mejor es un histograma

```python
%matplotlib inline
```

```python
precios.hist('aumento')
```

Como vemos, la mayoría está entre alrededor de -35 y 75. Entonces vamos a seleccionar sólo las filas donde `aumento` esté entre esos valores. 

Para eso usamos `.loc`.  Una de las posibilidad de `.loc` es recibir un array booleano de igual largo que el dataframe (misma cantidad de filas), seleccionando sólo aquellas filas donde el índice sea `True`. 

Por ejemplo consideremos el siguiente dataframe

```python
data = pd.DataFrame({'a': [1, 2, 3], 'b': ['a', 'b', 'c']})
```

Puedo seleccionar la primera y ultima fila con valores indizando con un array booleano asi

```python
data.loc[[True, False, True]]
```

Por otro lado, cuando comparamos una columna contra un valor, el resultado es el array booleano con `True` para aquellas filas en que la condición es verdadera y `False` cuando no. 

```python
data['a'] > 1
```

Si juntamos ambos conceptos, podemos seleccionar filas que cumplan una condición

```python
data.loc[data['a'] > 1]
```

Además, las condiciones se pueden combinar con operadores logicos 

```python
(data['a'] > 1) & (data['b'] == 'c') 
```

Entonces, sabemos como filtrar el rango de aumentos que deseamos

```python
precios_ = precios.loc[(precios['aumento'] > -35) &  (precios['aumento'] < 75)]
precios_.hist('aumento')
```

Ahora podemos ver otros aumentos más representativos para analizar en conjunto

```python
precios_.sort_values('aumento', ascending=False)[:50]
```

Hasta ahora, estamos considerando las filas, es decir el aumento de un producto en **una sucursal** en particular.
Si quisieramos evaluar el promedio (`.mean()`) o la mediana (`.median()`) de aumentos en cualquier sucursal, entonces podemos hacer un **agrupamiento por productos**. De esta manera, podemos calcular el aumento entre todas las filas que tengan el mismo identificador de producto, sin considerar la sucursal. 

```python
aumentos_promedio_por_producto =  precios_.groupby('producto_id').aumento.mean().to_frame('aumento')
```

`.to_frame()` sirve para convertir una serie en un dataframe de una sola columna, de manera de facilitar cruces y operaciones de la manera que ya las conocemos. 

```python
mas_aumentaron = aumentos_promedio_por_producto.sort_values('aumento', ascending=False)[:50]
mas_aumentaron
```

Si quisieramos ver detalles de cuales son estos productos, podemos cruzar con el dataset de productos

Un detalle con este dataset es que tiene "id" duplicados.

```python
productos.shape
```

```python
productos.drop_duplicates('id').shape
```

Por otro lado, en el dataset de la izquierda, los identificadores de productos ya no son una columna
sino que es *el indice*. Entonces en vez de usar `left_on` usamos directamente `left_index=True`

```python
pd.merge(mas_aumentaron, productos.drop_duplicates('id')[['id', 'nombre', 'marca']], left_index=True, right_on='id').sort_values('aumento', ascending=False)
```

Ahora extenmos el análisis para ver aumentos por marcas

```python
marca_producto = productos = pd.read_csv('../proyecto_integrador/productos.csv', usecols=['id', 'marca']).drop_duplicates('id')
marca_producto.head()
```

En este caso, la columna a cruzar es el codigo, que se llama `producto_id` en la izquierda e `id` en la derecha

```python
precios_con_marca = pd.merge(precios_, marca_producto, left_on='producto_id', right_on='id')
```

```python
precios_con_marca.head()
```

Repetimos el agrupamiento, pero esta vez por marca

```python
precios_con_marca.groupby('marca').aumento.mean().to_frame('aumento').sort_values('aumento', ascending=False)[:100] 
```
