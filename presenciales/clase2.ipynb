# Y ahora analicemos la inflación de mayo

```python
import pandas as pd
```

```python
precios_abril = pd.read_csv('../proyecto_integrador/precios_abril20.csv')
precios_abril.head()
```

```python
precios_mayo = pd.read_csv('../proyecto_integrador/precios_mayo22.csv')
precios_mayo.head()
```

```python
sucursales = pd.read_csv('../proyecto_integrador/sucursales.csv')
sucursales.head()
```

```python
cadenas = pd.read_csv('../proyecto_integrador/cadenas.csv')
cadenas.head()
```

```python
productos = pd.read_csv('../proyecto_integrador/productos.csv')
productos.head()
```

```python
precios = pd.merge(precios_abril, precios_mayo, on=['sucursal_id', 'producto_id'])
precios.head()
```

```python
precios = pd.merge(precios_abril, precios_mayo, on=['sucursal_id', 'producto_id'], suffixes=('_abril', '_mayo'))
precios.head()
```

```python
precios['aumento'] = (precios['precio_mayo'] - precios['precio_abril']) * 100 / precios['precio_abril']
precios.head()
```

```python
precios.sort_values('aumento', ascending=False)[:50]
```

```python
%matplotlib inline
```

```python
precios.hist('aumento')
```

```python
precios_ = precios.loc[(precios['aumento'] > -35) &  (precios['aumento'] < 75)]
precios_.hist('aumento')
```
